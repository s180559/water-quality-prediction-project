# Step 4: Model Selection and Training

Split the data into training and testing sets, balance the dataset, and select models for training.

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Reset the indices of X_train and y_train
X_train = X_train.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)

X_train_balanced, y_train_balanced = resample(X_train[y_train == 0], y_train[y_train == 0],
                                              n_samples=X_train[y_train == 1].shape[0], random_state=42)

X_train_balanced = pd.concat([X_train_balanced, X_train[y_train == 1]])
y_train_balanced = pd.concat([y_train_balanced, y_train[y_train == 1]])

classifiers = {
    "Logistic Regression": LogisticRegression(random_state=42),
    "Support Vector Machine": SVC(random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42)
}

best_accuracy = 0.0
best_classifier = None

for name, classifier in classifiers.items():
    classifier.fit(X_train_balanced, y_train_balanced)
    y_pred = classifier.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_classifier = name
    
    print(f"{name} Accuracy: {accuracy:.2f}")
```

# Step-7 Calculate WQI index for each row in the dataset

```python
import pandas as pd

# Load the dataset and select relevant columns
data = pd.read_csv('ground_water_quality_2020_pre .csv')
water_parameters = ['pH', 'E.C', 'TDS', 'CO3', 'HCO3', 'Cl', 'F', 'NO3 ', 'SO4', 'Na', 'K', 'Ca', 'Mg', 'T.H', 'SAR', 'RSC  meq  / L']

# Define the parameter weights (replace with actual weights)
parameter_weights = {
    'pH': 0.1,
    'E.C': 0.15,
    'TDS': 0.2,
    'CO3': 0.05,
    'HCO3': 0.08,
    'Cl': 0.07,
    'F': 0.05,
    'NO3 ': 0.03,
    'SO4': 0.05,
    'Na': 0.06,
    'K': 0.04,
    'Ca': 0.06,
    'Mg': 0.07,
    'T.H': 0.08,
    'SAR': 0.07,
    'RSC  meq  / L': 0.02
}

# Calculate sub-indices for each parameter
sub_indices = {}
for parameter in water_parameters:
    sub_indices[parameter] = data[parameter] * parameter_weights.get(parameter, 0)

# Calculate the overall WQI using a simple averaging method
data['WQI'] = sum(sub_indices.values())

# Assign labels based on WQI threshold
threshold = 50  # Adjust the threshold as needed
data['Classification'] = data['WQI'].apply(lambda x: 'MR' if x >= threshold else 'PS')
```

# Step-10 Anomaly detection is finding the contaminations

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
import matplotlib.pyplot as plt

# Load your water quality dataset
# Replace 'water_quality.csv' with your dataset file
data = pd.read_csv('ground_water_quality_2020_pre .csv')
# Select the columns you want to use for anomaly detection
selected_columns = ['pH', 'E.C', 'TDS', 'CO3', 'HCO3', 'Cl', 'F', 'NO3 ', 'SO4', 'Na', 'K', 'Ca', 'Mg', 'T.H']
X = data[selected_columns]

# Initialize the Isolation Forest model
isolation_forest = IsolationForest(contamination=0.05, random_state=42)

# Initialize the One-Class SVM model
one_class_svm = OneClassSVM(nu=0.05)

# Fit the models to the data
isolation_forest.fit(X)
one_class_svm.fit(X)

# Predict anomalies using the models
isolation_forest_predictions = isolation_forest.predict(X)
one_class_svm_predictions = one_class_svm.predict(X)

# Add the anomaly predictions as new columns in the original dataset
data['isolation_forest_anomaly'] = isolation_forest_predictions
data['one_class_svm_anomaly'] = one_class_svm_predictions

# Visualize anomalies
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.scatter(data.index, data['pH'], c=data['isolation_forest_anomaly'], cmap='coolwarm', marker='o')
plt.xlabel('Index')
plt.ylabel('pH Value')
plt.title('Isolation Forest Anomaly Detection')
plt.colorbar(label='Anomaly')

plt.subplot(1, 2, 2)
plt.scatter(data.index, data['pH'], c=data['one_class_svm_anomaly'], cmap='coolwarm', marker='o')
plt.xlabel('Index')
plt.ylabel('pH Value')
plt.title('One-Class SVM Anomaly Detection')
plt.colorbar(label='Anomaly')

plt.tight_layout()
plt.show()

# Display the data points flagged as anomalies for each method
isolation_forest_anomalies = data[data['isolation_forest_anomaly'] == -1]
one_class_svm_anomalies = data[data['one_class_svm_anomaly'] == -1]

print("Detected Anomalies using Isolation Forest:")
print(isolation_forest_anomalies)

print("\nDetected Anomalies using One-Class SVM:")
print(one_class_svm_anomalies)
```
